```python
plt.figure(figsize = (12,6))
sns.countplot(x='Genre', data=df)

plt.xticks(rotation=90);
```

```python
plt.figure(figsize = (12,6))
plt.bar(  df["Year"].value_counts().index,df["Year"].value_counts())
plt.title("Most games produced in a year")
plt.show()
```

```python
plt.figure(figsize = (12,6))
plt.bar(df["Platform"].value_counts().index, df["Platform"].value_counts())
plt.title("Most Games produced in Specific  Platform")
plt.show()
```

```python
plt.figure(figsize = (12,6))
plt.bar(df["Genre"].value_counts().index, df["Genre"].value_counts(),color ='m')
plt.title("Most Games produced in Specific Gaming Platform")
plt.show()
```
